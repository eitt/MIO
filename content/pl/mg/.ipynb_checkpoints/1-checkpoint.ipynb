## ¬øC√≥mo resolver un problema de programaci√≥n lineal?

De manera general, todos los problemas de optimizaci√≥n pueden ser estructurados en cinco pasos, estos parten desde la delimitaci√≥n del problema o situaci√≥n a mejorar, su descripci√≥n mediante funciones matem√°ticas, el hallazgo de posibles soluciones y la selecci√≥n de una soluci√≥n final, que puede ser √≥ptima o no dependiendo del problema. ### Definir las variables de decisi√≥n

Es fundamental una vez descrito el problema a solucionar determinar el tipo de variables. Seg√∫n las variables a usar (binarias, enteras, y/o cont√≠nuas [(Ver la secci√≥n de modelado matem√°tico)](/mm/1/01)), es necesario aplicar diferentes m√©todos de soluci√≥n, adem√°s, dependiendo del fen√≥meno a modelar algunas variables son m√°s pr√°cticas. Por ejemplo, se acostumbra a utilizar variables binarias para representar las diferentes asignaciones, a modo de ejemplo, la variable $x_{o,m}$ toma el valor de 1 si al operario Pedro trabaja o es asignado a la m√°quina 4. En este caso $x_{o,m}$ es una matriz que combina todos los posibles operarios (supongamos cinco: $\{Alonso, Beatriz, Carmen, Diego, Efren\}$), con todas las m√°quinas disponibles, a modo de ejemplo consideremos cuatro m√°quinas:  $\{m_1, m_2, m_3, m_4\}$, por tanto, $x_{o,m}$ es un conjunto de 20 variables.

En cuanto a las varaibles enteras y cont√≠nuas, estas son afines en su uso y s√∫ implicaci√≥n pr√°ctica para modelar es si la cantidad puede ser fraccionada o no. Por ejemplo, se puede suponer que se desea determinar la cantidad necesaria de impresoras 3D requeridas para fabricar unos repuestos. Si se considera comprar las m√°quinas, lo m√°s recomendado es usar variables enteras ya que no puedo comprar m√°quinas a medias; por otro lado, si el problema implica el alquiler de m√°quinas y estas puedan alquilarse por tiempo parcial, es recomendado trabajar con variables cont√≠nuas ya que puedo alquilar 5.5 m√°quinas, es decir, cinco m√°quinas tiempo completo y una m√°quina a medio tiempo.

Una vez definidas las variables se debe tener presente sus unidades y c√≥mo est√°s pueden variar en cantidad, tiempo o proporci√≥n. En cuanto a la cantidad, √©sta va de la mano con el tipo de variable; respecto al tiempo, es importante tener presente si la variable abarca todo el horizonte de planeaci√≥n o, si esta cambia seg√∫n el periodo. Retomando el ejemplo de la variable $x_{o,m}$, √©sta no parece indicar que cambie en el tiempo, es decir, Pedro trabajar√° en la m√°quina 4 durante el periodo completo, por el contrario, si se est√° asignando el operario a la m√°quina teniendo en cuenta la demanda mensual y se desea hacer una programaci√≥n anual,  problemente es necesario redefinir la variable con otro √≠ndice: $x_{o,m,t}$, donde $t$ es el periodo de tiempo que dura la asignaci√≥n, para este ejemplo, doce meses. Por tanto, $x_{o,m,t}$ es un conjunto de 240 variables de decisi√≥n $cantidad=5*4*12=240$. 

En cuanto a la proporci√≥n, al ser varaibles lineales su cambio tendr√° un impacto, por ejemplo, si no es lo mismo que Pedro trabaje en la m√°quina 4 que en la m√°quina 1 debido a que √©l es m√°s proficiente con la m√°quina 4 que con la 1, entonces si estoy analizando el tiempo que se demora en c√°da m√°quina se espera que sea mayor en la m√°quina 1 que en la m√°quina 4, por el contrario, si se est√° analizando el desperdicio derivado de la operaci√≥n, se espera que sea menor en la m√°quina 4 que en la 1. Teniendo esto presente, se debe verificar c√≥mo el cambio en el valor de cada variable de decisi√≥n puede afectar elementos como los costos, ingresos, satisfacci√≥n, demanda, requerimientos y dem√°s, elementos que ser√°n abordados m√°s adelante.

### Construir la funci√≥n objetivo

Una vez definidas las variables es necesario determinar su impacto en una m√©trica de calidad, es decir, una vez tomo una decisi√≥n √©sta genera un impacto en un objetivo de inter√©s. Si se est√° produciendo implantes impresos en 3D, tal vez el enfoque se relacione con consumo de material, costos de energ√≠a, valor de la mano de obra, ingresos, impacto ambiental, y as√≠ sucesivamente. De manera general, los objetivos en *PL* se pueden agrupar en dos grandes categor√≠as: el primero, objetivos de maximizar, esta categor√≠a abarca situaciones como ingresos, utilidad, satisfacci√≥n, felicidad, redimientos, eficiencia, es decir, todos los escenarios donde se desea aumentar el valor resultante; en contra parte, el segundo son los objetivos de minimizar, √©stos contienen situaciones como costos, dolor, movimientos, desperdicios, deserci√≥n y dem√°s; son todas esas situaciones donde se busca disminuir el valor.

Teniendo en cuenta el tipo de variables y la informaci√≥n disponible, se asigna un peso a cada variable; este peso indica c√≥mo impacta la decisi√≥n en la funci√≥n objetivo. Ejemplificando, si se est√° comprando insumo pl√°stico para una impresora 3D, y seespera consumir entre 100 kilogramos a 500 kilogramos, donde la variable $x_{inv}$, es la cantidad de inventario que se espera comprar, se puede hallar un costo asociado a la compra, por ejemplo, suponer que cada kilogramo de material pl√°stico cuenta 1000, por tanto, en la funci√≥n objetivo debe aparecer dentro de la suma de costos dicho peso: $FO=1000x_{inv}$.

En a√±adidura, si se est√° considenrando comprar el material por cada unidad de tiempo durante los doce meses, es espera halla un sub√≠ndice extra relacionado con el periodo: $x_{inv,t}$, que indicar√≠a la cantidad de material comprado durante el periodo $t$, en consecuencia, dentro de la funci√≥n objetivo se espera encontrar todos esos costos asociados:

\begin{equation*}
FO=1000x_{inv,1}+1000x_{inv,2},\dots,1000x_{inv,12}
\end{equation*}

Cabe resaltar que al ser costos constantes, es posible reescribir la funci√≥n factorizando:

\begin{equation*}
FO=1000(x_{inv,1}+x_{inv,2},\dots,x_{inv,12})
\end{equation*}

Sin embargo, si los costos cambian seg√∫n cada periodo (por ejemplo los *black friday* u eventos similares), no ser√≠a posible factorizar y, por tanto, suponiendo costos entre 1000 y 1500, se obtendr√≠a la sigueinte funci√≥n:

\begin{equation*}
FO=1500x_{inv,1}+1115x_{inv,2},\dots,1300x_{inv,12}
\end{equation*}

Con el prop√≥sito de hacer m√°s compacta la notaci√≥n, se suele describir los costos como conjuntos que pueden ser vectores o matrices seg√∫n la cantidad de √≠ndices que tenga la variable que est√°n premultiplicando. en este ejemplo, es posible definir un costo $c$ relacionado con el inventario en el tiempo, es decir, $c_{inv,t}$. Teniendo en cuenta lo anterior, y sabiendo que $c_{inv,t}$ puede ser un vector con los siguientes valores: 

$c_{inv,t}=\{1500,1115,1235,1224,1469,1000,1500,1450,1369,1232,1187,1300\}$

Por tanto, la funci√≥n objetivo puede expresarse de la siguiente manera:

\begin{equation*}
FO=c_{inv,1}x_{inv,1}+c_{inv,2}x_{inv,2},\dots,c_{inv,12}x_{inv,12}
\end{equation*}

Dicha estructura lineal crece a la medida que se consideran m√°s variables o nuevos √≠ndices, por consiguiente, con el fin de facilitar la escritura se suele condensar mediante el operador de sumatoria como se muestra a continuaci√≥n.

\begin{equation*}
FO=\sum_{t=1}^{12} c_{inv,t}x_{inv,t}
\end{equation*}

No obstante, teniendo en cuenta que se tienen dos vectores de igual tama√±o, uno asociado a los costos y otro relacionado con las variables de decisi√≥n. Otra manera de exponer la funci√≥n objetivo es como el producto punto entre ambos vectores:

\begin{equation*}
FO=c_{inv,t}.x_{inv,t}
\end{equation*}

Para finalizar, si bien se ha ejemplificado la construcci√≥n de la funci√≥n objetivo con un caso de costos, el procedimiento es el mismo para los casos de beneficios. En consecuencia, se acostumbra a llamar carga al valor que multiplica dada variable de decisi√≥n con el fin de que contemple los escenarios donde aumentar o asignar un valor a la variable impacte tanto positiva como negativamente la FO.

### Determinar el criterio de decisi√≥n
De manera paralela a la construcci√≥n de la funci√≥n objetivo se determina el criterio de decisi√≥n, √©ste se agrupa en dos categor√≠as que van de la mano con los grupos de funciones objetivo: M√°ximizar, es decir, asignar valores a las variables de decisi√≥n de tal forma que se encuentre el mayor valor posible, este criterio se aplica para la categor√≠a de maximizar. En contra parte, est√° el criterio de minimizar que busca el m√≠nimo valor posible de FO.

### Construir el espacio de b√∫squeda

El espacio de b√∫squeda es el conjunto de todos los posibles valores que pueden tomar las variables de decisi√≥n. Es necesario indicar que comunmente existen restricciones que se deben cumplir y √©stas acotan dicho espacio. Retomando el ejemplo de x_{inv,t}, existe la posibilidad que la cantidad de material que se compre durante cada mes no deba superar la capacidad de almacenamiento, por tanto, el valor de x_{inv,t} no puede superar la capacidad en el instante $t$. Si se supone que se cuenta con una capacidad variable entre una y dos toneladas por mes, a modo de ejemplo se puede tener el siguiente sigueinte vector de capacidades de almacenamiento $cap=\{1,1,2,1,1,1,2,1,1,1,2,2\} [ton]$, en consecuencia se tienen las siguientes doce restricciones asociadas a no comprar m√°s inventario del que se pueda almacenar por mes:

\begin{equation*}
r1:x_{inv,1}<=1\\
r2:x_{inv,2}<=1\\
r3:x_{inv,3}<=2\\
r4:x_{inv,4}<=1\\
r5:x_{inv,5}<=1\\
r6:x_{inv,6}<=1\\
r7:x_{inv,7}<=2\\
r8:x_{inv,8}<=1\\
r9:x_{inv,9}<=1\\
r10:x_{inv,10}<=1\\
r11:x_{inv,11}<=2\\
r12:x_{inv,12}<=2
\end{equation*}

Adem√°s, existen situaciones en las cuales la suma de varias variables deben cumplir con ciertas restricciones. Por ejemplo, se suponga que la cantidad total de inventario por a√±o no debe superar las trece toneladas, ni ser menor a ocho toneladas. Dicha informaci√≥n genera dos nuevas restricciones:

\begin{equation*}

r14:\sum_{t=1}^{12} x_{inv,t}>=8
\end{equation*}

As√≠ mismo, existen escenarios donde se contemplan de manera simult√°nea varias variables pero no en su totalidad, suponga ahora que durante los primeros tres meses la cantidad comprada debe ser inferior a seis toneladas. La nueva restricci√≥n ser√°:

\begin{equation*}
r15:x_{inv,1}+x_{inv,2}+x_{inv,3}<=6
\end{equation*}

Ahora bien, teniendo en cuenta que en un PL suelen existir m√∫ltiples restricciones que no consideran todas las variables, es posible incluirlas premultiplic√°ndolas por el valor de $0$. A modo de ejemplo, es posible reescribir la ecuaci√≥n 15 de la siguiente manera:

\begin{equation*}
r15:x_{inv,1}+x_{inv,2}+x_{inv,3}+0x_{inv,4}+0x_{inv,5}+0x_{inv,6}+0x_{inv,7}+0x_{inv,8}+0x_{inv,9}+0x_{inv,10}+0x_{inv,11}+0x_{inv,12}<=6
\end{equation*}

Teniendo en cuenta lo anterior, es posible aplicar el mismo procedimiento a todas las restricciones. Para el ejemplo se tendr√≠an 15 restricciones con 12 variables:

\begin{equation*}
r1:x_{inv,1}+0x_{inv,2}+0x_{inv,3}+\dots+0x_{inv,12}<=1\\
r2:0x_{inv,1}+x_{inv,2}+0x_{inv,3}+\dots+0x_{inv,12}<=1\\
r3:0x_{inv,1}+0x_{inv,2}+x_{inv,3}+\dots+0x_{inv,12}<=2\\
\vdots \:\:\:\:\:\:\:\:\:\: \:\:\:\:\:\:\:\:\:\:\:\:\:\:\ddots  \:\:\:\:\:\:\:\:\:\: \:\:\:\:\:\:\:\:\:\:\:\:\:\:\vdots \\
r15:x_{inv,1}+x_{inv,2}+x_{inv,3}+\dots+0x_{inv,12}<=6
\end{equation*}

Dicho lo anterior, al tener presentes todas las variables es posible reescribir el conjunto de restricciones como la operaci√≥n entre una matriz de cargas (los valores que premutiplican a las variables de decisi√≥n), y el vector columna de las variables de decisi√≥n. Esta multiplicaci√≥n entre matrices relaciona la cantidad de restricciones ($m$), con el n√∫mero de variables ($n$), y genera un vector columna de $m$ elementos. En el ejemplo descrito anteriormente, $m=15, n=12$, los coeficientes de la matriz de restricciones son:

A = \begin{bmatrix} 
    1 & 0 & 0 & \dots & 0 \\
    0 & 1 & 0 & \dots & 0 \\
    0 & 0 & 1 & \dots & 0 \\
    \vdots & \ddots & \\
    1 & 1 & 1 & \dots & 0 
    \end{bmatrix}

Y el vector de variables es:

x = \begin{bmatrix} 
    x_{inv,1} \\
    x_{inv,2} \\
    x_{inv,3} \\
    \vdots \\
    x_{inv,12}
    \end{bmatrix}

Entonces, para el ejemplo:

**\begin{equation*}
A*x<=b
\end{equation*}**

Donde **$b$** es el vector columna de decisiones:

b = \begin{bmatrix} 
    b_{1}=1 \\
    b_{2}=1 \\
    b_{3}=2 \\
    \vdots \\
    b_{15}=6
    \end{bmatrix}
    
Lo cual se estructura de manera muy similar a la forma est√°ndar de un PL, con la diferencia de presentar restricciones de desigualdad en vez de igualdad; no obstante, toda desigualdad puede ser reescrita como una igualdad mediante el uso de variables auxiliares. esta estrategia ser√° abordada m√°s delante en la secci√≥n donde se explica el m√©todo de soluci√≥n Simplex.

A modo de cierre, m√∫ltiples problemas de programaci√≥n lineal pueden ser modelados siguiendo la estructura de restricciones previamente mostrada. Ahora bien, diversas situaciones requieren aplicar distintas estructuras como el uso de variables auxiliares para restricciones excluyentes, el el manejo de variables que relacionen asignaciones con magnitudes. Dichas aproximaciones puden ser profundizadas en el trabajo de Williams {% cite williams2013model%}.

### Construir el espacio de soluciones

Una vez definido el conjunto que contiene a posibles soluciones teniendo en cuenta las restricciones del problema, se generan las soluciones alternativas. Donde hay dos variables de decisi√≥n $\{x_1,x_2\}$ y una funci√≥n objetivo $FO$, √©sta transforma el conjunto de decisiones $R^2$ en un conjunto de posibles soluciones $R$, es decir, por cada posible decisi√≥n $\{x_1,x_2\}$  hay asociada mediante una funci√≥n lineal una soluci√≥n $z_i$, la figura indica el proceso para $k$ posibles soluciones. En la pr√°ctica, se busca seleccionar la mejor soluci√≥n (m√°ximo o m√≠nimo) de la recta $FO$
