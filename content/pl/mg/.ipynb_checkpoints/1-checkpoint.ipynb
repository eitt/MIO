## Introducci√≥n

El m√©todo gr√°fico para la resoluci√≥n de problemas de programaci√≥n lineal es generalmente utilizado de manera introductoria. Principalmente debido a que √©ste se aplica cuando existen s√≥lo dos variables de decisi√≥n, condici√≥n que no es com√∫n en la mayor√≠a de problemas; sin embargo, es a partir del an√°lisis gr√°fico que es posible describir el [M√©todo Simplex](/pl/ms/1) que permite la resoluci√≥n de problemas con m√°s de dos variables.

En esencia, el m√©todo gr√°fico es un an√°lisis del espacio de b√∫squeda y su correspondiente funci√≥n objetivo, con la ventaja de poder representar de manera simult√°nea tres dimensiones que corresponden a los ejes de las abscisas ($x_1,x_2$) y el eje ordenado de la funci√≥n objetivo, com√∫nmente descrito con la variable $z$. Por consiguiente, el m√©todo gr√°fico es el an√°lisis de crecimiento o decrecimiento del plano $z$ que depende de las variables $x_1,x_2$, es decir, $z=f(x_1,x_2)$, donde esta relaci√≥n se puede reescribir de la siguiente manera: $z-a_1x_1-a_2x_2=0$, que la respectiva ecuaci√≥n de un plano, donde $a_1 y a_2$ son los coeficientes que pre multiplican a las variables de decisi√≥n.

De manera general, los problemas de programaci√≥n lineal implican variables de decisi√≥n que compiten entre s√≠ por el uso de recursos limitados, por ejemplo, para la construcci√≥n de implantes impresos en 3D, una compa√±√≠a se especializa en el desarrollo de dos tipos de implantes: 

1. Implantes dentales.
2. Implantes √≥seos.

Con el fin de simplificar, se puede suponer que s√≥lo se fabrica una referencia de cada tipo de implante. Como ambos implantes utilizan la misma materia prima, es necesario decidir cu√°nto de cada tipo de implante se va a fabricar considerando elementos como la demanda de los productos, los recursos en com√∫n (maquinaria, empleados, materia prima, etc.), ingresos percibidos, y dem√°s elementos o consideraciones que permitan describir el problema a resolver.

Esta situaci√≥n (diversos productos que compiten entre s√≠ por recursos) se denomina frecuentemente mezcla de productos, y su estructura es ampliamente utilizada en diversos campos como: construcci√≥n, agricultura, metalurgia, servicios, entre otros. {% cite Heady1954%}


## Espacio de b√∫squeda en el m√©todo gr√°fico

Como se indica en los [cinco pasos para modelar un *PL*](/pl/intro), una vez definidas las variables, la funci√≥n objetivo con su respectivo criterio, es necesario construir el espacio de b√∫squeda, o espacio de restricciones. Para ello, se puede partir de la siguiente inecuaci√≥n: $ùëì(ùë•_1,ùë•_2)‚â§ùëè$ o $ùëì(ùë•_1,ùë•_2)‚â•ùëè$. 

Se puede demostrar que, en dos dimensiones, el conjunto de puntos que satisface una desigualdad lineal (o inecuaci√≥n) comprende o contiene los puntos en la recta $ùëì(ùë•_1,ùë•_2)=ùëè$, que define la desigualdad m√°s todos los puntos de un lado de la recta (puntos por encima en el caso de ser mayor o igual $"‚â•"$, o los puntos inferiores para el escenario de menor igual $"‚â§"$). Sin embargo, en algunos casos no es tan intuitivo determinar qu√© lado de la recta corresponde a la desigualdad, por tanto, se elige un punto $ùëÉ$ arbitrario y a partir de este se eval√∫a la condici√≥n de la restricci√≥n.

### Restricciones de igualdad
Normalmente un PL de dos variables, cuyas variables de decisi√≥n corresponden a $ùë•_1$  y $ùë•_2$, presenta una forma similar a la  siguiente restricci√≥n:

\begin{equation*}
-2ùë•_1+1.5ùë•_2=6
\end{equation*}

Donde los valores de $-2$ y de $1.5$ que pre multiplican a las variables $ùë•_1,ùë•_2$ son un ejemplo de las posibles realizaciones de la carga $c\in R$. Al observar la ecuaci√≥n que describe la restricci√≥n se encuentra que el mismo conjunto de puntos satisface $(ùë•_1,ùë•_2)$ satisface:

\begin{equation*}
1.5ùë•_2=6+2ùë•_1
\end{equation*}

De igual forma satisface:

\begin{equation*}
ùë•_2=4‚àí\frac{2}{1.5}ùë•_1=>ùë•_2=4‚àí\frac{4}{3}ùë•_1
\end{equation*}

o:

\begin{equation*}
ùë•_1=\frac{1.5ùë•_2-6}{-2}=>ùë•_1=\frac{-3}{4}ùë•_2+3 
\end{equation*}

Seg√∫n la variable que se desee expresar como dependiente.

Ahora bien, esta relaci√≥n al estar en $R^2$ puede ser representada de manera gr√°fica. En *R* es posible realizar gr√°ficas de funciones, para ello son necesarios los siguientes pasos:

1. Definir la funci√≥n.
2. Determinar el dominio de la variable en el eje de las abscisas.
3. Utilizar un paquete *package* de gr√°ficos.

Por lo general, como buena pr√°ctica se cargan los paquetes al principio de cada c√≥digo. Como tal se trabajar√° con el paquete [ggplot2](https://ggplot2.tidyverse.org/) √©ste permite generar de manera flexible diferentes tipos de gr√°ficos, para ahondar m√°s se pueden consultar *R for data science* {% cite wickham2016r%}, *ggplot2: elegant graphics for data analysis* {% cite wickham2016ggplot2%}, y *R graphics cookbook* {% cite chang2018r%}. A continuaci√≥n se muestra la manera en la cual se puede graficar la restricci√≥n de ejemplo $r.1=ùë•_2=4‚àí\frac{4}{3}ùë•_1$

```r
#1. Se carga el paquete
library(ggplot2)
#2. Se define la funci√≥n relacionada con la restricci√≥n 1 (r.1)
r.1 <- function(x) 4-4/3*x
#3. Se declaran los l√≠mites:
#L√≠mite inferior (l.i)
l.i <- 0 #Las variables normalmente son restrictas y mayores o iguales a cero
#L√≠mite superior (l.s)
l.s <- 10 #L√≠mite arbitrario
```
Una vez definida la estructura de la funci√≥n, se procede a graficar utilizando el paquete `ggplot2`

```r
#Se acostumbra a almacenar la gr√°fica en una variable denominada p (por gr√°fico en ingl√©s)
#Se genera un conjunto de datos artificial para generar la estructura base
p <- ggplot(data = data.frame(x = 0), mapping = aes(x = x))
#Se agrega la funci√≥n definida previamente, incluyendo los l√≠mites.
p + stat_function(fun = fun.1) + xlim(l.i,l.s)
```
En resumen, la ejecuci√≥n del c√≥digo y su respectiva gr√°fica se indica a continuaci√≥n.

````r
#1. Se carga el paquete
library(ggplot2)
#2. Se define la funci√≥n relacionada con la restricci√≥n 1 (r.1)
r.1 <- function(x) {4-4/3*x}
#3. Se declaran los l√≠mites:
#L√≠mite inferior (l.i)
l.i <- 0 #Las variables normalmente son restrictas y mayores o iguales a cero
#L√≠mite superior (l.s)
l.s <- 10 #L√≠mite arbitrario

#Se acostumbra a almacenar la gr√°fica en una variable denominada p (por gr√°fico en ingl√©s)
#Se genera un conjunto de datos artificial para generar la estructura base
p <- ggplot(data = data.frame(x = 0), mapping = aes(x = x))
#Se agrega la funci√≥n definida previamente, incluyendo los l√≠mites.
p + stat_function(fun = r.1) + xlim(l.i,l.s)
````

Como tal, la funci√≥n resultante genera una gr√°fica de l√≠nea recta, donde a medida que aumenta el valor de la variable en el eje de las abscisas, disminuye el de la variable ordenada. Como por defecto el paquete de `ggplot2` genera un gr√°fico con ejes denominados $x$ y $y$, a continuaci√≥n, se agrega una estructura de c√≥digo que permite personalizar el gr√°fico. El par√°metro `colour` permite cambiar el color del gr√°fico, mientras `labs` configura el t√≠tulo de los ejes para $x_1$  `labs(x = "x1")`, y $x_2$ `labs(y = "x2")` respectivamente.

```r
p + stat_function(fun = r.1, colour = "red")+ xlim(l.i,l.s) + labs(x = "x1") + labs(y = "x2")
```

De la recta se puede concluir que existen infinitos puntos $(x_1,x_2)$ que satisfacen la restricci√≥n, en consecuencia, la decisi√≥n ser√° cu√°l par de puntos optimiza la funci√≥n FO.

En caso de querer visualizar otra restricci√≥n de manera simult√°nea, es necesario definir funciones extras, una por cada nueva restricci√≥n, y a√±adirlas al gr√°fico. Con el fin de estandarizar el esquema de colores a utilizar se trabajar√° con la paleta Dark2 [RColorBrewer](http://colorbrewer2.org/) desarrollada por Neuwirth y otros {% cite neuwirth2014colorbrewer%}. Para utilizar los colores es necesario cargar el paquete mediante la instancia `library(RColorBrewer)`la visualizaci√≥n de los colores y su respectiva codificaci√≥n hexadecimal es:

````r
library(RColorBrewer)
# Visualizaci√≥n de los colores
display.brewer.pal(n = 8, name = 'Dark2')
# Color en notaci√≥n Hexadecimal
brewer.pal(n = 8, name = "Dark2")
````

Para finalizar, se generar√° a modo de ejemplo otra restricci√≥n con el fin de indicar c√≥mo graficar m√∫ltiples restricciones utilizando el c√≥digo de `ggplot2`. Se considerar√° la siguiente restricci√≥n:

\begin{equation*}
r2:ùë•_2=-1+ùë•_1
\end{equation*}

Repitiendo el proceso, es necesario definir la segunda funci√≥n con sus respectivos colores, adem√°s. se debe definir un rango para cada una de las restricciones, por ejemplo, para la restricci√≥n 1 es: `rango.1 <- data.frame(x=c(l.i, l.s), Functions = factor(1))`, donde `x=c(l.i, l.s)` indica el dominio de la variable $x_1$, y `Functions = factor(1))` permite agrupar m√∫ltiples funciones como un factor. En cuanto a los colores, la paleta se asigna manualmente mediante el c√≥digo mostrado anteriormente `scale_colour_manual(values = c("#1B9E77", "#D95F02")` (los dos primeros colores). Para finalizar, `labels = c("Restricci√≥n 1", "Restricci√≥n 2")`y `labs(x = "x1") + labs(y = "x2")` generan las leyendas y t√≠tulos de los ejes respectivamente.


````r
#Primera restricci√≥n
r.1 <- function(x) 4-4/3*x
#Segunda restricci√≥n
r.2 <- function(x) -1+x
#Rangos para cada restricci√≥n
rango.1 <- data.frame(x=c(l.i, l.s), Functions = factor(1))
rango.2 <- data.frame(x=c(l.i, l.s), Functions = factor(2))
#Generaci√≥n del gr√°fico
ggplot(NULL, aes(x=x, colour=Functions)) +
  stat_function(data = plot.range1, fun = r.1) +
  stat_function(data = plot.range2, fun = r.2) +
  scale_colour_manual(values = c("#1B9E77", "#D95F02"), labels = c("Restricci√≥n 1", "Restricci√≥n 2")) +
  labs(x = "x1") + labs(y = "x2")
````

Es importante resaltar que, cuando se trabajan con restricciones de igualdad es posible que no exista una soluci√≥n, para el caso mostrado anteriormente, existe una √∫nica soluci√≥n y corresponde con el punto donde ambas soluciones se cruzan, es decir, el punto que cumple con ambas restricciones.

### Restricciones de desigualdad



````r
#Primera restricci√≥n
r.1 <- function(x) 4-4/3*x
#Segunda restricci√≥n
r.2 <- function(x) -1+x
#Rangos para cada restricci√≥n
rango.1 <- data.frame(x=c(l.i, l.s), Functions = factor(1))
rango.2 <- data.frame(x=c(l.i, l.s), Functions = factor(2))
#Generaci√≥n del gr√°fico
ggplot(NULL, aes(x=x, colour=Functions)) +
  stat_function(data = plot.range1, fun = r.1,geom="area", fill="#1B9E77", alpha=0.3) +
  stat_function(data = plot.range2, fun = r.2,geom="area", fill="#D95F02", alpha=0.3) +
  scale_colour_manual(values = c("#1B9E77", "#D95F02"), labels = c("Restricci√≥n 1", "Restricci√≥n 2")) +
  labs(x = "x1") + labs(y = "x2")
````
